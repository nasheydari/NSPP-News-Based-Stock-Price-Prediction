```python
import pynytimes
import pandas as pd
```

```python
from pynytimes import NYTAPI
```

```python
from my_api_info import get_nytimes_key
```

```python
nytapi = NYTAPI(get_nytimes_key(),
                parse_dates=True)
```

```python
from datetime import datetime
```

```python
results = nytapi.article_search(query="stock",
                                results=365,
                                dates={'begin':datetime(2023,1,1),
                                       'end':datetime(2023,12,30)})
```

```python
NYdata = pd.DataFrame(results)
NYdata['pub_date'] = pd.to_datetime(NYdata['pub_date'])
```

```python
NYdata.head()
```

```python
sp_price=pd.read_csv("Download Data - INDEX_US_S&P US_SPX.csv", parse_dates=['Date'])
```

```python
sp_price.set_index('Date', inplace=True)
```

```python
sp_price.sort_index()
```

```python
NYdata['Date']=NYdata['pub_date'].dt.date
```

```python
NYdata.set_index('Date', inplace=True)
```

```python
NYdata.head()
```

```python
NYdata = NYdata[~NYdata.index.duplicated(keep='first')]
```

```python
NYdata.sort_index()
```

```python
NYdata.to_csv('NYdata.csv', index=True)
```

```python
Data_combined = pd.merge(NYdata, sp_price, left_index=True, right_index=True, how='outer')
```

```python
Data_combined['abstract'][1]
```

```python
Data_combined.to_csv('NYdata_SP.csv', index=True)
```

```python
Data_combined_dict={}
Data_combined_dict['train']=Data_combined.loc[Data_combined.index.month<11]
Data_combined_dict['validation']=Data_combined.loc[Data_combined.index.month==11]
Data_combined_dict['test']=Data_combined.loc[Data_combined.index.month>11]
```

```python
len(Data_combined_dict['validate'])
```

```python
from datasets import Dataset, DatasetDict
```

```python
datasets_dict = {key: Dataset.from_pandas(df) for key, df in Data_combined_dict.items()}

# Create a DatasetDict object
dataset_dict = DatasetDict(datasets_dict)
```
